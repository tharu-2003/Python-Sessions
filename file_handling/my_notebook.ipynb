## =========== TEXT ===========

## ------------ Read => read() with "r" ------------

```python
file_1 = open("sample_1.txt" , "r")

content = file_1.read()

print(content)

file_1.close()
```

```python
with open("sample_1.txt" , "r") as file :

    content = file.read()
    print(content)
```

## ------------ Read a line => readline() ------------

```python
with open("sample_1.txt" , "r") as file :

    line_1 = file.readline()
    line_1 = line_1.strip()
    print(line_1)
    line_2 = file.readline()
    print(line_2)
```

```python
with open("sample_1.txt" , "r") as file :


    while file:
        
        line = file.readline().strip()
        
        if not line:
            break

        print(line)
```

## ------------ Read lines => readlines() ------------

```python
with open("sample_1.txt" , "r") as file :
    lines = file.readlines()
    print(lines)
```

```python
with open("sample_1.txt" , "r") as file :
    lines = file.readlines()
    
    for line in lines:
       print(line.strip())
```

## ------------New Write => write() with "w" ------------

```python
with open("sample_1.txt" , "w") as file :

    file.write("we are writing to a file")
```

## ------------Append Write => write() with "a" ------------

```python
with open("sample_1.txt" , "a") as file :

    file.write("we are writing to a file")
```

```python
with open("sample_2.txt" , "a") as file :
    file.write("we are writing to a file")
```

## ------------ Write lines => writelines() ------------

```python
with open("sample_3.txt" , "w") as file :

    lines = ["Line One content\n", "Line Two content\n"]
    file.writelines(lines)
```

## =========== JSON ===========

### ---------- Read as dictionary => json.load() ----------

```python
import json

with open("my_json.json", "r") as json_file:

    content = json.load(json_file)
print(content)
print(type(content))
```

### ---------- Wirte as dictionary => json.dump() ----------

```python
import json

with open("sample.json", "w") as json_file:

    human = {
        'name': 'John', 
        'age': 12,
        'city': 'Galle'
    }

    json.dump(human, json_file, indent=4) # (indent=4) -> meka nisa pahapata lassanata hedenawa
```

## =========== CSV ===========

### ---------- Read csv file => csv.reader() ----------

```python
import csv

with open("person_data.csv", "r", newline='') as csv_file:

    table = csv.reader(csv_file)

    # print(content)
    for row in table:
        print(row)
```

### ---------- Read csv file as Dictionary => csv.DictReader() ----------

```python
import csv

csv_file_path = "person_data.csv"

with open(csv_file_path, "r") as csv_file:

    table = csv.DictReader(csv_file)

    for dictionary in table:
        print(dictionary)
```

### ---------- Write csv file as Dictionary List => DictWriter() ----------

```python
import csv

data = [
    {"Name": "Tharusha", "Age": 22, "City": "Galle"}, 
    {"Name": "Sachini", "Age": 21, "City": "Diyathalawa"},
    {"Name": "Anjana", "Age": 22, "City": "Elpitiya"}
]

filednames = ["Name", "Age", "City"]

with open("sample_output.csv", "w", newline='') as csv_file:

    csv_writer = csv.DictWriter(csv_file, fieldnames = filednames)

    csv_writer.writeheader()

    csv_writer.writerows(data)

print(f"CSV data has been written to {csv_file_path}")
```
