## ========== Polymorphism ========== 

```python
class Shape:
    def render(self):
        print ("Rendering Shape")

class Circle(Shape):
    def render(self):
        print("Rendering Circle")

class Square(Shape):
    def render(self):
        print("Rendering Square")

shape1 = Circle()
shape2 = Square()

shape1.render()  # Output: Rendering Circle
shape2.render()  # Output: Rendering Square
```

## ========== Encapsulation ==========

### ----------- Protected ( _name )----------- 

```python
class Square:

    def __init__(self, name):
        self._name = name

    def _area(self, side_len):  
        return side_len ** 2
    
square_1 = Square("Square")
print(square_1._area(5), " ", square_1._name)
```

```python
class Person:
    def __init__(self):
        self._name = "Tharusha"

class Employee(Person):
   def display(self):
       print(self._name) # OK - accessed in subclasses

e = Employee()
print(e._name) # Technically allowed , but not recommand
```

### ----------- Private ( __name )----------- 

```python
class Square:

    def __init__(self, name):
        self.__name = name

    def __area(self, side_len):  
        return side_len ** 2
    
    def setName(self, name):
        self.__name = name

    def getName(self):
        return self.__name
    
    def getArea(self,side_len):
        return self.__area(side_len)

    
square_1 = Square("MySquer")

square_1.setName("MySquer")
print(square_1.getName())
print(square_1.getArea(5))
```
