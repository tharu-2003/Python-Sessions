## ---------------- Operaters with NumPy ----------------

```python
import numpy as np

arr_1 = np.array([5,9,12,11])
arr_2 = np.array([3,9,4,8])

sum_arr = arr_1 + arr_2
dic_arr = arr_1 - arr_2
mul_arr = arr_1 * arr_2
div_arr = arr_1 / arr_2
divi_arr = arr_1 // arr_2
di_arr = arr_1 % arr_2

print(sum_arr)
print(dic_arr)
print(mul_arr)
print(div_arr)
print(divi_arr)
print(di_arr)
```

## ---------------- NumPy pre defined methods ----------------

```python
import numpy as np

arr = np.array([16,9,25,64])

# ---------------- sqrt( ) ----------------
sqrt_arr = np.sqrt(arr)
print(sqrt_arr)

# ---------------- max( ) ----------------
max_value = np.max(arr)
print(max_value)

# ---------------- min( ) ----------------
min_value = np.min(arr)
print(min_value)

# ---------------- sum( ) ----------------
sum_value = np.sum(arr)
print(sum_value)

# ---------------- mean( ) ----------------
mean_value = np.mean(arr)
print(mean_value)

```

```python
# ---------------------------------- Q4 -------------------------------------

import pandas as pd

df = pd.DataFrame({
    "Product": ['Laptop' , 'Mouse' , 'Keyboard' , 'Monitor' , 'Headset'],
    "UnitsSold": [12, 50, 34, 20, 40],
    "Price": [120000 , 1500 , 3500 , 25000 , 6000],
    "Discount": [0.10 , 0.05 , 0.08 , 0.12 , 0.07]
})

final_price = df['Price'] - (df['Price'] * df['Discount']) 

total_revenue = final_price * df['UnitsSold']

df['FinalPrice'] = final_price
df['TotalRevenue'] = total_revenue


print(final_price)
print()
print(total_revenue)
print()
print(df)
print()
print(df[ (df['TotalRevenue'] > 500000)])
print()
print('Total Revenue of All Products ' , df['TotalRevenue'].sum())

```
