# ==== requests.post(url, data) ====

```python
import requests

data = {
    'username': 'user',
    'password': 'pass'
}

url = 'https://httpbin.org/post'

response = requests.post(url, data)

if response.status_code == 200:
    print("Login Success")
    print(response.text)
else:
    print('Failed to Login :' , response.status_code)
```

## ------- URL Parameters  -------

### ----- params = params -----

```python
import requests

url = 'https://jsonplaceholder.typicode.com/posts'

params = {
    'user_id': 1,
    'id': 2
}

response = requests.get(url, params=params)

if response.status_code == 200:
    print(response.text)
else:
    print('Failed to Get Data :' , response.status_code)
```

### ----- auth=(username, password) -----

```python
import requests

url = 'https://httpbin.org/get'

username = 'user'
password = 'pass'

response = requests.get(url, auth=(username, password))

if response.status_code == 200:
    print(response.text)
else:
    print('Failed Authentication :' , response.status_code)
```

# =========== Beautiful Soup Library ===========

```python
from bs4 import BeautifulSoup

html_doc = """<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>
<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>
<p class="story">...</p>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

print(soup.title)
print(soup.title.text)
print('-------------')
print(soup.p.contents)

print(soup.a)
print(soup.p)

print('-------------')
print(soup.find('a'))

print('-------------')
print(soup.find_all('a'))
print('-------------')
print(soup.find_all('p'))
```

# ========= Weather data =========

```python
# ------------------------------- Q2 ------------------------------
import requests

base_url = "https://api.open-meteo.com/v1/forecast"

params = {
    'latitude': 6.075,
    'longitude': 80.21,
    'current': 'temperature_2m,wind_speed_10m',
    # 'hourly': 'temperature_2m,relative_humidity_2m,wind_speed_10m'
}   

response = requests.get(base_url, params=params)

if response.status_code == 200:
    print(response.headers['Content-Type'])
    print(response.headers['date'])
    print("Successfuly feched the page")
    wather_data = response.json()
    
    if wather_data and 'current' in wather_data:
        print(wather_data)
        current_wather = wather_data.get('current')

        temperature = current_wather.get('temperature_2m')
        wind_speed = current_wather.get('wind_speed_10m')

        print(f"Temperature : {temperature} C" )
        print(f"Wind Speed : {wind_speed} km/h", )

    else:
        print('Weather data is Not Available')
else:
    print('Failed to Feched data :' , response.status_code)



```
