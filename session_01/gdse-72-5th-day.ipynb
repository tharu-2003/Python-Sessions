```python
# --------------------------Q2---------------------------

class Car:
    wheels = 4

    def drive_to(self, destination):
        print(f"Driving to {destination} on {self.wheels} wheels")

    def show_wheel_count(self):
        print(f"The car runs on {self.wheels} wheels")

car_1 = Car()

car_1.drive_to("Colombo")
car_1.show_wheel_count()
```

```python
# ------------------------Q3--------------------------

class Car:
    wheels = 4

    def __init__(self, brand, color):
        
        self.brand = brand
        self.color = color

    def show_wheel_count(self):
        return self.wheels
    
car_2 = Car("BMW", "Red")

print(f"{car_2.color} color {car_2.brand} has {car_2.show_wheel_count()} wheels")
```

```python
# ---------------------------Q4-----------------------

from math import pi

class Circle:
    
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return  pi * (self.radius ** 2)
    
    def circumference(self):
        return 2 * pi * self.radius
    
    def display_info(self):
        print(f"circle with radius {self.radius} has area {self.area()} and cricumference {self.circumference()}")

circle_1 = Circle(7)

circle_1.display_info()
```

## Inheritance


```python
class Animal:

    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} speak")

class Dog(Animal):

    def __init__(self, name, color):
        super().__init__(name)
        self.color = color
    
    def speak(self):
        print(f"{self.name} say woof.It is {self.color}") # method overriding

dog_1 = Dog("bauwa", "white")
dog_1.speak()
```

```python
from math import pi

class Shape:
    
    def __init__(self, name):
        self.name = name

    def area(self):
        print(f"Area calculation not defined for {self.name}")

class Square(Shape):

    def __init__(self, name, length):
        super().__init__(name)
        self.length = length

    def area(self):
        print(f"{self.name} Area calculation is {self.length * self.length}")

class Circle(Shape):

    def __init__(self, name, radius):
        super().__init__(name)
        self.radius = radius

    def area(self):

        print(f"{self.name} Area calculation is {pi * (self.radius**2)}")

square = Square("Square", 10)
square.area()

circle = Circle("Circle", 7)
circle.area()
```
