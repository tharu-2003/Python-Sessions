```python
my_list_2 = [10, 11, 15.5, "A", "S", True]
my_list_3 = [10, 12, 8 , 9, 22, 21, 5, 11]

print(my_list_2)
print(type(my_list_2))

print(my_list_2.index("A")) #index view

print(my_list_2[1:5])

print(my_list_2[:3])
print(my_list_2[3:]) 

print(my_list_3[0:7:3]) #print patterns

my_list_3[2:5] = [200, 300, 400] #data reasign to multipla segment
print(my_list_3)

my_list_3.insert(1,77) #data insert to one segment
print(my_list_3)

my_list_3.append(66) #add data to the last segment
print(my_list_3)

my_list_3.extend([777, 555, 444]) #add datas to the last segment
print(my_list_3)
```
